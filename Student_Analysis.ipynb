```python
import numpy as n
```

```python
# Columns: [Age, Math Marks, Science Marks]
data = n.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])

shape= data.shape
print("the shape is",shape)
```

```python
average_age=n.mean(data[:,0])
print("average age is",average_age)
```

```python
maths_marks=data[:,1]
print("all maths marks:",maths_marks)
```

```python
highest_science_marks=n.max(data[:,2])
print("highest marks achieved in science is:",highest_science_marks)
```

```python
student_info=data[:,1]>90
detail=data[student_info]
print("detail of a student who scored more than 90:",detail)
```

```python
data2=data.copy()
data2[:,1]+=5
print("the new updated data is:\n",data2)
```

```python
age=data[:,0]<19
count=n.sum(age)
print(f"there are total {count} students who are below 19")
```

```python
average_marks_subjects=n.mean(data[:,1:3],axis=0)
print("the average marks in each subject:")
average_marks_subjects
```

```python
score=data[:,1:3]>=80
score_detail=data[score.all(axis=1)]
print("the detail of the students are here:",score_detail)
```

```python
new_data=data.copy()
marks=new_data[:,2]<75
new_data[marks,2]=0
print("the new data now:\n",new_data)
```
